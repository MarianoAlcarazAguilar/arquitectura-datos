## Ejemplo de uso

```python
from nasa_api_controler import NasaControler
```

```python
# Funcionalidad básica
controlador = NasaControler()
controlador.load_image('2023-01-02') # cargamos una imagen
controlador.display_image() # Mostramos la imagen cargada
```

```python
# Guardar una imagen
controlador = NasaControler()
controlador.load_image('2023-01-01')
controlador.save_image() # Por default la guarda en la misma carpeta donde se ejecuta este comando con el nombre de la fecha buscada
```

```python
# Guardar una imagen en una carpeta de imágenes con el nombre default
controlador = NasaControler(images_folder='images/') # Le decimos dónde queremos que guarde las imágenes
controlador.load_image('2023-01-01')
controlador.save_image() # La guarda en la carpeta usando el nombre default

# El mismo resultado se puede obtener de la siguiente manera
controlador = NasaControler()
controlador.load_image('2023-01-01')
controlador.save_image(location='images/2023-01-01.png') # En este caso, es necesario especificar de nuevo el nombre a usar
```

```python
# Guardar una imagen con un nombre específico
controlador = NasaControler()
controlador.load_image('2023-01-01') # cargamos una imagen
controlador.save_image(location='images/mi_imagen.png')
```
